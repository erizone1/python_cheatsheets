## Two dimensional datasets

```python
y = np.random.standard_normal((20,2)).cumsum(axis=0) # generate 2 dimensional dataset
```

```python
plt.figure(figsize=(10,6))
plt.plot(y,lw=1.5)
plt.plot(y,'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A simple plot')
```

```python
plt.figure(figsize=(10,6))
plt.plot(y[:,0],lw=1.5, label='1st')           # defines labels for subsets of the data
plt.plot(y[:,1],lw=1.5, label='2nd')
plt.plot(y,'ro')
plt.legend(loc=0)                              # place labels in the 'best' location
plt.xlabel('index')
plt.ylabel('value')
plt.title('A simple plot');
```

## Approach with different scaling on y axis is dimensions of the 2 plots do not match

```python
y[:,0] = y[:,0]*100
# by using the plt.subplots() function one gets direct access to the underlying plotting objects. It allows for example to generate a second subplot that shares the x-axis with the first one
fig, ax1 = plt.subplots()            # defines figures and axis objects

plt.plot(y[:,0], 'b', lw=1.5, label='1st')
plt.plot(y[:,0],'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A simple plot')
ax2=ax1.twinx() # Create a new Axes with an invisible x-axis and an independent y-axis positioned opposite to the original one (i.e. at right). The x-axis autoscale setting will be inherited from the original Axes.  To ensure that the tick marks of both y-axes align, see `~matplotlib.ticker.LinearLocator`.
plt.plot(y[:,1], 'g', lw=1.5, label='2nd')
plt.plot(y[:,1],'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd');

```

```python
# Next cosider the case of two separate subplots
plt.figure(figsize=(10,6))
plt.subplot(211)  # another way to write numrows=2 numcols=1 fignum=2
plt.plot(y[:,0],lw=1.5,label='1st')
plt.plot(y[:,0],'ro')
plt.ylabel('value');
plt.title('Another simple plot')
plt.legend(loc=0)
plt.subplot(212)
plt.plot(y[:,1],'g', lw=1.5,label='2nd')
plt.plot(y[:,1],'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value');
```

```python
plt.figure(figsize=(10,6))
plt.subplot(121)    # another way to write numrows=1 numcols=2 fignum=2
plt.plot(y[:,0],lw=1.5,label='1st')
plt.plot(y[:,0],'ro')
 
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('1st data set')
plt.subplot(122)
plt.bar(np.arange(len(y)),y[:,1],width=0.5, color='g', label='2nd') #creates the bar subplot
plt.legend(loc=0)
plt.xlabel('index')
plt.title('2nd data set')
```

### Mathematically inspired plot
plots function and highlights graphically the area under the function (integral).
The integral to be illustrated is $\int_{a}^{b} f(x) \,dx$ with $ f(x) = \frac{1}{2}\exp(x) + 1 \$
This demonstrates that mathplotlib seamlessly handles LaTeX typesetting.

#### Define function and Integral

```python
def f(x):
    return(0.5*np.exp(x)+1)
a,b = 0.5,1.5
x=np.linspace(0,2)
y=f(x)
Ix=np.linspace(a,b)
Iy=f(Ix)
verts=[(a,0)] + list(zip(Ix,Iy))+[(b,0)]
```

```python
verts
```

 #### plot itself

```python
from matplotlib.patches import Polygon
fig, ax = plt.subplots(figsize=(10,6))
plt.plot(x,y,'b',linewidth=2)
plt.ylim(bottom=0) # defines lower limit for ordinate axis
poly=Polygon(verts,facecolor='0.7',edgecolor='0.5')
ax.add_patch(poly)
plt.text(0.5*(a + b),1, r'$\int_a^b f(x)\mathrm{d}x$', fontsize=20, horizontalalignment='center')
plt.figtext(0.9,0.075,'$x$')
plt.figtext(0.075,0.9,'$f(x)$')
ax.set_xticks((a,b))
ax.set_xticklabels(('$a$', '$b$'))
ax.set_yticks([f(a), f(b)])
ax.set_yticklabels(('$f(a)$', '$f(b)$'))
```
